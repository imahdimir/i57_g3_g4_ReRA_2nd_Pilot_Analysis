```python
import pandas as pd
from i7_hot_reload.hot_reload import hot_reload_a_module_from_cwd


b_000_env = hot_reload_a_module_from_cwd('b_000_env')

from b_000_env import Directory
from b_000_env import FilePath
from b_000_env import Constants


d = Directory()
fp = FilePath()
c = Constants()
```

```python
# Read Pages times data
df_pg_t_0 = pd.read_excel(fp.cln_page_times)
print(df_pg_t_0.shape)

df_pg_t_0 = df_pg_t_0.drop_duplicates()

df_pg_t_0.shape
```

```python
# sort by participant code and page index
df_pg_t = df_pg_t_0.sort_values(by = [c.participant_code , c.page_index])
```

```python
# keep only first and last page for each participant code, to get total time on the survey
df_first_last_page = df_pg_t.groupby(c.participant_code ,
                                     group_keys = False).apply(lambda g : g.loc[
    [g[c.page_index].idxmin() , g[c.page_index].idxmax()]])
```

```python
# calculate total time on the survey
cols_2_keep = [c.participant_code , c.page_index , c.epoch_time_completed]
df_total_time = df_first_last_page[cols_2_keep].copy()

df_total_time[c.total_track_time_sec] = \
    df_first_last_page.groupby(c.participant_code)[
        c.epoch_time_completed].diff()

df_total_time = df_total_time.dropna()
df_total_time.shape
```

```python
# add track name for each participant code
df_cln_all_apps = pd.read_excel(fp.clean_all_apps)

df_cln_all_apps = df_cln_all_apps[
    [c.participant_code , c.participant_sni_treatment , c.participant_track ,
     c.participant_payoff , c.mturk_id]]
```

```python
_comp = 'Comparison'
_tran = 'Transitivity'

_l = 'Large'
_si = 'SmallIncentivized'
_sn = 'SmallNonincentivized'

values = {
        (_l , _comp)  : (c.cl , 'CL') ,
        (_si , _comp) : (c.csi , 'CSI') ,
        (_sn , _comp) : (c.csu , 'CSU') ,
        (_l , _tran)  : (c.tl , 'TL') ,
        (_si , _tran) : (c.tsi , 'TSI') ,
        (_sn , _tran) : (c.tsu , 'TSU') ,
        }

for k , v in values.items() :
    df_cln_all_apps.loc[
        (df_cln_all_apps[c.participant_sni_treatment] == k[0]) & (
                df_cln_all_apps[c.participant_track] == k[1]) , [c.track ,
                                                                 c.track_abbreviation]] = v

df_cln_all_apps.to_excel(fp.m3__participant_code_track_map , index = False)
```

```python
df_cln_all_apps_1 = df_cln_all_apps.drop(columns = [
        c.participant_sni_treatment , c.participant_track])

df_total_time_1 = pd.merge(df_total_time ,
                           df_cln_all_apps_1 ,
                           on = c.participant_code)
```

```python
df_total_time_1[c.total_track_time_min] = df_total_time_1[
                                              c.total_track_time_sec] / 60
```
