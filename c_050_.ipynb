```python
import pandas as pd
from i7_hot_reload.hot_reload import hot_reload_a_module_from_cwd


b_000_env = hot_reload_a_module_from_cwd('b_000_env')

from b_000_env import Directory
from b_000_env import FilePath
from b_000_env import Constants


d = Directory()
fp = FilePath()
c = Constants()

import warnings


warnings.filterwarnings("ignore")
```

```python
df_track_times = pd.read_excel(fp.m4__total_track_time)
df_track_times.head()
```

```python
df_all = df_track_times[c.total_track_time_sec].describe().to_frame().T
df_all
```

```python
df_by_track = df_track_times.groupby(c.track_abbreviation)[
    c.total_track_time_sec].describe()
df_by_track
```

```python
df_table = pd.concat([df_by_track , df_all])
df_table
```

```python
def fmt_as_hh_mm(s) :
    hours , remainder = divmod(s , 3600)
    minutes , seconds = divmod(remainder , 60)
    if seconds > 30 :
        minutes += 1
    return '{:02}:{:02}'.format(int(hours) , int(minutes))
```

```python
df = df_table
df.iloc[: , 1 :] = df.iloc[: , 1 :].applymap(lambda x : fmt_as_hh_mm(x))
df
```

```python
df['count'] = df['count'].astype('Int32')

df = df.rename(columns = {
        'count' : 'N' ,
        'mean'  : 'Mean' ,
        'std'   : 'STD' ,
        'min'   : 'Min' ,
        '25%'   : 'Q1' ,
        '50%'   : 'Median' ,
        '75%'   : 'Q3' ,
        'max'   : 'Max'
        })

df = df.astype('string')

df
```
